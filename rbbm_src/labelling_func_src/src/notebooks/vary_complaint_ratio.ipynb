```python
import pandas as pd
import glob
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
```

```python
dfs = []
for f in glob.glob('change_complaint_ratio*'):
    df=pd.read_csv(f)
    dfs.append(df)
```

```python
df = pd.concat(dfs)
df['user_input_cnt']=df['num_complaints']+df['confirmation_cnt']
l = ['strat',
 'runtime',
 'avg_tree_size_increase',
 'user_input_cnt',
 'num_complaints',
 'confirmation_cnt',
 'global_accuracy',
 'fix_rate',
 'confirm_preserve_rate',
 'new_global_accuracy',
 'prev_signaled_cnt',
 'new_signaled_cnt']
df[l].sort_values(['user_input_cnt','strat'])
```

```python
df.groupby(['strat','user_input_cnt','num_complaints']).mean()
```
