```python
from gensim.test.utils import datapath
from gensim import utils
import os

cwd = os. getcwd()

class MyCorpus:
    """An iterator that yields sentences (lists of str)."""

    def __iter__(self):
        corpus_path = datapath(f'{cwd}/spam_corpus.txt')
        for line in open(corpus_path):
            # assume there's one document per line, tokens separated by whitespace
            yield utils.simple_preprocess(line)
```

```python
import gensim.models

sentences = MyCorpus()
model = gensim.models.Word2Vec(sentences=sentences)
```

```python
for index, word in enumerate(model.wv.index_to_key):
    if index == 10:
        break
    print(f"word #{index}/{len(model.wv.index_to_key)} is {word}")
```

```python
words_ued_by_lfs = ["check","out","my","subscribe","http","please","plz","song","shakira","google","money","$","the"]
```

```python
example_text = """
Top three Shakira songs (my choice) <br /><br />1- Waka Waka (it&#39;s time for Africa)<br /><br />2- Can&#39;t remember to forget you <br /><br />3- Empire<br /><br />Like this comment if u like Shakira
"""

d = {}
for w in example_text.split():
    d[w] = []
    
```

```python
for w in d:
    for wu in words_ued_by_lfs:
        try:
            d[w].append([wu, wv.similarity(w, wu)])
        except:
            continue
```

```python
for k,v in d.items():
    d[k]=sorted(d[k], key=lambda x: x[1], reverse=True)
```
