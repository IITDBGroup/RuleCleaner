```python
def line_plots_legends_only(color_assignment_dict):
    # Create a figure and axis
    fig, ax = plt.subplots()

    # Remove background grid
    ax.set_axis_off()

    # Create custom legend entries
    legend_entries = []

    for strat, color in color_assignment_dict.items():
            legend_entries.append(mpatches.Patch(facecolor=color, edgecolor='black', label=strat))
#             legend_entries.append(mlines.Line2D([], [], color=color, label=strat, markersize=30))
    # Use Seaborn to add a legend
#     ax.legend(handles=legend_entries, loc='center', title='Strat')
    ax.legend(handles=legend_entries, loc='center', ncol=4,     handlelength=1.2)
    plt.savefig(f"graph_pdfs/lf_del_thresh_legend_only.pdf", format="pdf",bbox_inches='tight')
    plt.show()
    plt.close(fig)
    
custom_palette = ['#6ca5ce', '#ffb675', '#60b760', '#d72c2d']
strats = ['hit rate.', 'global acc.', 'new global acc.', 'rel. size']

color_dict = dict(zip(strats, custom_palette))

line_plots_legends_only(color_dict)
```

```python
# Filter the data for different gt_bad_func_cnt values
gt_bad_func_cnt_values = list(range(15, 27,3)) # Add more values as needed

# Create a single figure for all subplots
fig, axs = plt.subplots(len(gt_bad_func_cnt_values), 1, figsize=(8, 8))
legend_labels = ['Hit Rate', 'Global Accuracy', 'New Global Accuracy', 'relative_total_size']

# Plot each subplot
for i, cnt in enumerate(gt_bad_func_cnt_values):
    # Filter data based on gt_bad_func_cnt
    filtered_data = df_deletion[df_deletion['gt_bad_func_cnt'] == cnt]

    # Plot hit_rate, global_accuracy, and new_global_accuracy against deletion_absolute_threshold
    axs[i].plot(filtered_data['deletion_absolute_threshold'], filtered_data['hit_rate'], label='Hit Rate')
    axs[i].plot(filtered_data['deletion_absolute_threshold'], filtered_data['global_accuracy'], label='Global Accuracy')
    axs[i].plot(filtered_data['deletion_absolute_threshold'], filtered_data['new_global_accuracy'], label='New Global Accuracy')
    axs[i].plot(filtered_data['deletion_absolute_threshold'], filtered_data['relative_total_size'], label='Relative Total Size')
    # Set axis labels and title
    axs[i].set_xlabel('Deletion Absolute Threshold')
    axs[i].set_ylabel('Values')
    axs[i].set_title(f'gt_bad_func_cnt = {cnt}')

# Add a single legend to the combined figure
fig.legend(legend_labels, loc='upper right',bbox_to_anchor=(1.2, 0.85))

# Adjust the layout to prevent overlapping
plt.tight_layout()

# Show the combined figure
plt.show()
```
