```python
import glob
import pandas as pd
```

```python
dicts = []
for fi in glob.glob('raw_dataset/*/ham/*.txt'):
#     print(fi)
    with open(fi, 'r', encoding='utf-8',errors='ignore') as f:
        string = f.read()
    d = {'text': string, 'label': 0}
    dicts.append(d)
ham_df = pd.DataFrame(data=dicts)

dicts = []
for fi in glob.glob('raw_dataset/*/spam/*.txt'):
#     print(fi)
    with open(fi, 'r', encoding='utf-8',errors='ignore') as f:
        string = f.read()
    d = {'text': string, 'label': 1}
    dicts.append(d)
spam_df = pd.DataFrame(data=dicts)
```
