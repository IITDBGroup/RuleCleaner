```python
lf_rules = []
```

```python
class l_func:
    def __init__(self, name, regex, label):
        self.name=name
        self.regex=regex
        self.label=label
    def evaluate(self, x):
        return self.label if re.search(self.regex, x, flags=re.I) else ABSTAIN
```

```python
for r in regex_rules:
    reg_text = r.regex
    label = r.label
#     @lfunc_dec(name=r.rid, words_used=[reg_text, label])
    rule=l_func(name=r.rid, regex=reg_text, label=label)
#         return label if re.search(reg_text, x, flags=re.I) else ABSTAIN
    lf_rules.append(rule)
```
